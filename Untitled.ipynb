```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# WARNING: TAKES ALOT OF TIME
data = pd.read_csv('data.csv')
data.head()
```

```python
print(data.dtypes)
#data['Id'] = pd.to_numeric(data['Id'])
print(data.dtypes)
data.shape
```

```python
# WHAT TO DO WITH NULL VALUES?

print(data.isnull().sum())
print(data[data.duplicated()].shape[0])

data['Company'] = data['Company'].fillna('Unknown')
data['Location'] = data['Location'].fillna('Unknown')
data['Area'] = data['Area'].fillna('Unknown')
data['Classification'] = data['Classification'].fillna('Unknown')
data['SubClassification'] = data['SubClassification'].fillna('Unknown')
data['Requirement'] = data['Requirement'].fillna('Unknown')
data['FullDescription'] = data['FullDescription'].fillna('Unknown')
data['JobType'] = data['JobType'].fillna('Unknown')

print(data.isnull().sum())
print(data[data.duplicated()].shape[0])
```

```python
# PART 2 - DATA ANALYSIS AND INTERPRETATION
# PERFORM EXPLORATORY DATA ANALYSIS
test = data.groupby('JobType').mean()
print(test)
plt.plot(test)
```
